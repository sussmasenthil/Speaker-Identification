```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Audio_features01.csv')
df = pd.DataFrame(df)
df.head()
pred = df.drop('ID',axis = 1)
pred.head()
```

```python
from sklearn.preprocessing import LabelEncoder
ss=StandardScaler()
enc = LabelEncoder()
target = enc.fit_transform(df.ID)
pred.Gender = enc.fit_transform(df.Gender)
Scaled_x = pd.DataFrame(ss.fit_transform(pred))
```

```python
pca = PCA()
pca.fit_transform(Scaled_x)
pca.explained_variance_
```

```python
plt.semilogy(pca.explained_variance_,'--o')
plt.title('Scree Plot')
```
